### Question 1: Write a Python program to count the number of elements in a list without using len.

### Question 7: Write a Python program to find which number is palindrome in below given list .

- `sample input : `
    - `var = [134,121,155]`
- `sample output :`
    - `result : 134 is not a palindome , 121 is a palindrome `

```python
var=[134,121,155]
for i in var:
  a=str(i) # number converted to str
  b=a[::-1] # str reverse
  if a==b:
    print(f"palindrome:{i}")
  else:
    print(f"not a palindrome:{i}")
```

```python
# Please write The solution Here for above question 7:
var=int(input()) #[134,121,155]
temp=var
c=0
while var>0:
  res=var%10
  c=(c*10)+res
  var=var//10
if temp==c:
  print(f"palindrome:{temp}")
else:
  print(f"not a palindrome:{temp}")
```

### Question 8: Write a Python program to find the common elements between two lists.

- `sample input : `
    - `var1 = [1,2,3,4,10]`
    - `var2 = [10,11,12,14]`
- `sample output :`
    - `result : 10`

```python
# Please write The solution Here for above question 8:
var1=[1,2,3,4,10]
var2=[10,11,12,14]
a=set(var1)
b=set(var2)
c=list(a.intersection(b)) #c=[10]
print(f"result: {c[0]}")
```

### Question 9: Write a Python program to remove the duplicates from a list

- `sample input : `
    - `var = [10,11,12,10,12,24]`
- `sample output :`
    - `result : [10,11,12,24]`

```python
var=[10,11,12,10,12,24]
res=[]
for i in var:
  if i not in res:
    res.append(i)
print(f"result: {res}")
```
